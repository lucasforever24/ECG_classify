```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
patients = os.listdir("data/new_data/pulmonary_sinus")
patients_dir = os.path.join("data/new_data/pulmonary_sinus", patients[26])
files = os.listdir(patients_dir)
files.sort()
```

```python
patients = os.listdir("data/new_data/aortic_sinus")
patients_dir = os.path.join("data/new_data/aortic_sinus", patients[20])
files = os.listdir(patients_dir)
files.sort()
```

```python
lines = []
for k in range(len(files)):
    with open(os.path.join(patients_dir, files[k])) as f:
        lead = f.readlines()
        lead = np.asarray(lead)
        lines.append(lead)
    
lines = np.asarray(lines).astype('float')
print(lines.shape)
```

```python
def plot_ecg(ecg):
    figure = plt.figure(figsize=(12, 8))
    axes1 = figure.add_subplot(4, 3, 1)
    axes2 = figure.add_subplot(4, 3, 2)
    axes3 = figure.add_subplot(4, 3, 3)
    axes4 = figure.add_subplot(4, 3, 4)
    axes5 = figure.add_subplot(4, 3, 5)
    axes6 = figure.add_subplot(4, 3, 6)
    axes7 = figure.add_subplot(4, 3, 7)
    axes8 = figure.add_subplot(4, 3, 8)
    axes9 = figure.add_subplot(4, 3, 9)
    axes10 = figure.add_subplot(4, 3, 10)
    axes11 = figure.add_subplot(4, 3, 11)
    axes12 = figure.add_subplot(4, 3, 12)
    
    axes1.plot(ecg[0])
    axes2.plot(ecg[1])
    axes3.plot(ecg[2])
    axes4.plot(ecg[3])
    axes5.plot(ecg[4])
    axes6.plot(ecg[5])
    axes7.plot(ecg[6])
    axes8.plot(ecg[7])
    axes9.plot(ecg[8])
    axes10.plot(ecg[9])
    axes11.plot(ecg[10])
    axes12.plot(ecg[11])
    
    figure.show()
```

```python
plot_ecg(lines[0:12])
```
