```python
import pickle
import os
import matplotlib.pyplot as plt
```

```python
result_log_path = "2020-07-27 20:55/result_log.p"

with open(result_log_path, 'rb') as f:
    result_dict = pickle.load(f)
    
print(result_dict.keys())
```

```python
train_losses = result_dict['train_losses']
train_counter = result_dict['train_counter']
val_losses = result_dict['val_losses']
val_counter = result_dict['val_counter']
val_accuracy = result_dict['val_accuracy']
test_losses = result_dict['test_losses']
test_accuracy = result_dict['test_accuracy']
```

```python
print(test_accuracy, test_losses)
```

```python
fig = plt.figure(figsize=(12, 8))
ax1 = fig.add_subplot(111)
ax1.plot(val_counter, val_losses, color='black')
# ax1.plot(val_counter, val_accuracy, color='black')

ax2 = ax1.twinx()
ax2.plot(val_counter, val_accuracy, 'r')

plt.show()
```
